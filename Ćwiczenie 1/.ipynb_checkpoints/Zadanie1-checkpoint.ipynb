```python
def ELU(x,alpha=0.01):
    return np.maximum(alpha*(np.exp(x)-1),x)
    
def elu_gradient(x, alpha=0.01):
    return np.where(x >= 0, 1, alpha * np.exp(x))
    
x = np.linspace(-5,5,100)    
y = ELU(x,0.01)

grad = elu_gradient(x,1.0)
                     
fig, ax = plt.subplots()
ax.plot(x, y, label="ELU")
ax.plot(x, grad, label="gradient")
ax.legend()
plt.show()
```
